## 1. Fórmulas de Conversión

Para convertir entre escalas de temperatura, utilizamos las siguientes relaciones matemáticas basadas en los puntos de congelación y ebullición del agua:

* **De Celsius a Fahrenheit:** multiplicamos por 1.8 (o 9/5) y sumamos 32.
    $$°F = (°C \times 1.8) + 32$$

* **De Fahrenheit a Celsius:** restamos 32 y dividimos el resultado entre 1.8.
    $$°C = \frac{°F - 32}{1.8}$$



---

## 2. Ejemplos Resueltos Paso a Paso

### Caso A: Convertir 25 °C a Fahrenheit (Temperatura ambiente cómoda)
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Multiplicar:** $25 \times 1.8 = 45$
4.  **Sumar 32:** $45 + 32 = 77$
* **Resultado:** 25 °C equivalen a **77 °F**.

### Caso B: Convertir 98.6 °F a Celsius (Temperatura corporal promedio)
1.  **Identificar la fórmula:** $°C = (°F - 32) / 1.8$
2.  **Sustituir el valor:** $°C = (98.6 - 32) / 1.8$
3.  **Restar 32:** $98.6 - 32 = 66.6$
4.  **Dividir entre 1.8:** $66.6 / 1.8 = 37$
* **Resultado:** 98.6 °F equivalen a **37 °C**.

## 3. Código en Python

Este programa es interactivo y utiliza funciones para separar la lógica de los cálculos de la interfaz del menú.

```python
def convertirCF():
    celsius = float(input("\nIngrese los grados Celsius a convertir: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertirFC():
    fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES DE TEMPERATURA ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")
        
        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida, por favor intente de nuevo.")

# Llamada inicial para arrancar el programa
if __name__ == "__main__":
    menu()
```

### Explicación breve del código:
* **`while True`**: Crea un bucle infinito para que el programa no se cierre después de una sola conversión.
* **`float()`**: Permite que el usuario ingrese números con decimales (importante en ciencia).
* **`f"{valor:.2f}"`**: Es un formato para que el resultado solo muestre 2 decimales y sea más fácil de leer.
* **`break`**: Rompe el ciclo y finaliza la ejecución cuando el usuario elige la opción 3.

````python
def convertirCF():
    celsius = float(input("\nIngrese los grados Celsius a convertir: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertirFC():
    fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES DE TEMPERATURA ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")

        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Karla Saenz 0133")
            break
        else:
            print("Opción no válida, por favor intente de nuevo.")

# Llamada inicial para arrancar el programa
if __name__ == "__main__":
    menu()
````
